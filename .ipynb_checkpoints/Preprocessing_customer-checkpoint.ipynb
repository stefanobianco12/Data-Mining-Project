```python
def iqr_values(s): 
    q1 = s.quantile(q = 0.25)
    q3 = s.quantile(q = 0.75)
    iqr = q3 - q1
    iqr_left = q1 - 1.5*iqr
    iqr_right = q3 + 1.5*iqr
    return iqr_left, iqr_right
```

```python
left_Nsessions, right_Nsessions= iqr_values(df.Nsessions)
df[(df.Nsessions > left_Nsessions) & (df.Nsessions < right_Nsessions)].Nsessions.plot.box()
plt.show()
```

```python
left_AvgItemSession, right_AvgItemSession= iqr_values(df.avgItemSession)
df[(df.avgItemSession > left_AvgItemSession) & (df.avgItemSession < right_AvgItemSession)].avgItemSession.plot.box()
plt.show()
```

```python
left_MeanTimeGap, right_MeanTimeGap= iqr_values(df.MeanTimeGap)
df[(df.MeanTimeGap > left_MeanTimeGap) & (df.MeanTimeGap < right_MeanTimeGap)].MeanTimeGap.plot.box()
plt.show()
```

```python
left_TotSpent, right_TotSpent= iqr_values(df.TotSpent)
df[(df.TotSpent > left_TotSpent) & (df.TotSpent < right_TotSpent)].TotSpent.plot.box()
plt.show()
```

```python
left_avgSpent, right_avgSpent= iqr_values(df.avgSpent)
df[(df.avgSpent > left_avgSpent) & (df.avgSpent < right_avgSpent)].avgSpent.plot.box()
plt.show()
```

```python
left_Min, right_Min= iqr_values(df.Min)
df[(df.Min > left_Min) & (df.Min < right_Min)].Min.plot.box()
plt.show()
```

```python
left_MinItem, right_MinItem= iqr_values(df.MinItem)
df[(df.MinItem > left_MinItem) & (df.MinItem < right_MinItem)].MinItem.plot.box()
plt.show()
```

```python
left_Max, right_Max= iqr_values(df.Max)
df[(df.Max > left_Max) & (df.Max < right_Max)].Max.plot.box()
plt.show()
```

```python
left_DistinctItem, right_DistinctItem= iqr_values(df.DistinctItem)
df[(df.DistinctItem > left_DistinctItem) & (df.DistinctItem < right_DistinctItem)].DistinctItem.plot.box()
plt.show()
```

```python
left_MaxItem, right_MaxItem= iqr_values(df.MaxItem)
df[(df.MaxItem > left_MaxItem) & (df.MaxItem < right_MaxItem)].MaxItem.plot.box()
plt.show()
```

```python
left_TotItem, right_TotItem= iqr_values(df.TotItem)
df[(df.TotItem > left_TotItem) & (df.TotItem < right_TotItem)].TotItem.plot.box()
plt.show()
```
