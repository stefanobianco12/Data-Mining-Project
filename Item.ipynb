```python
#Tot of item purchased ordered by "most purchased Item"
TotItemBought = df.groupby('StockCode', as_index=False)['Quantity'].sum()
TotItemBought.columns = ['StockCode', 'TotItemBought']
#TotItem.sort_values(by=['TotItem'], ascending=False)
TotItemBought
```

```python
#Number of distinct customer for Item
DistinctCust = df.groupby('StockCode')['Customer ID'].nunique().reset_index()
DistinctCust.columns = ['StockCode', 'DistinctCust']
DistinctCust
```

```python
#The number of distinct session where an Item was bought 
DistinctSession = df.groupby('StockCode')['Invoice'].nunique().reset_index()
DistinctSession.columns = ['StockCode', 'DistinctSession']
DistinctSession 
```

```python
#Verifico che 10002R sia stato acquistato in 3 sessioni dallo stesso Customer
mask=df['StockCode']=='10002R'
fii=df[mask]
print(fii)
```

```python
#Mean Item bought for Month
MeanItemMonth = df.groupby('StockCode', as_index=False)['Quantity'].sum()
MeanItemMonth.columns = ['StockCode', 'MeanItemMonth']
MeanItemMonth['MeanItemMonth'] = MeanItemMonth['MeanItemMonth']/36
MeanItemMonth
```

```python
#mean of item bought for session (totItem/nsess)
AvgItemSess =TotItemBought.merge(DistinctSession, on='StockCode').set_index('StockCode')
AvgItemSess['AvgItemSess']=AvgItemSess['TotItemBought']/AvgItemSess['DistinctSession']
AvgItemSess
```

```python
MeanTimeGap = df.groupby(['StockCode', 'InvoiceDate'], as_index=False)['Invoice'].count()
MeanTimeGap.columns = ['StockCode', 'InvoiceDate','MeanTimeGap']
MeanTimeGap
MeanTimeGap['InvoiceDate'] = pd.to_datetime(MeanTimeGap['InvoiceDate']).dt.date
```

```python
#ogni quanto Ã¨ stato acquistato un item (n giorni)
def gapTime(MeanTimeGap):
    mean_time_gap=[]
    for item in MeanTimeGap['StockCode'].unique():
        df_aux=MeanTimeGap[MeanTimeGap['StockCode']==item]
        df_aux=df_aux['InvoiceDate'].to_list()
        sum_diff_gap = 0
        for i in range(0, len(df_aux)-1):
            sum_diff_gap += (df_aux[i+1]-df_aux[i]).days

        if len(df_aux)-1 == 0:
            mean_time_gap.append(0)
        else:
            mean_time_gap.append(sum_diff_gap / (len(df_aux)-1))
    return mean_time_gap
```

```python
MeanTimeGap
```

```python
mean_time_gap=gapTime(MeanTimeGap)
```

```python
item=MeanTimeGap['StockCode'].unique()
data_tuples = list(zip(item,mean_time_gap))
MeanTime=pd.DataFrame(data_tuples, columns=['StockCode','MeanTimeGap'])
MeanTime
```
